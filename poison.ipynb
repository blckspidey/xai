```python
import numpy as np
import pandas as pd

def poison_labels(df, poison_percent):

    np.random.seed(42)

    df_poisoned = df.copy()

    n_samples = len(df)
    n_poison = int(poison_percent * n_samples)

    indices = np.random.choice(n_samples, n_poison, replace=False)

    for idx in indices:
        original = df_poisoned.loc[idx, "label"]

        possible_labels = [0,1,2,3]
        possible_labels.remove(original)

        new_label = np.random.choice(possible_labels)

        df_poisoned.loc[idx, "label"] = new_label

    return df_poisoned
```

```python
train_df = pd.read_csv("/content/TRAINING_NORMAL (1).csv")

train_5 = poison_labels(train_df, 0.05)
train_25 = poison_labels(train_df, 0.25)
train_50 = poison_labels(train_df, 0.50)
train_75 = poison_labels(train_df, 0.75)

train_5.to_csv("POISONED_5percent.csv", index=False)
train_25.to_csv("POISONED_25percent.csv", index=False)
train_50.to_csv("POISONED_50percent.csv", index=False)
train_75.to_csv("POISONED_75percent.csv", index=False)
```
